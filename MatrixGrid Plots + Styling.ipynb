## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


```python
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

### Heatmaps (heatmap())

```python
import seaborn as sn
import matplotlib.pyplot as plt

df  = sn.load_dataset('tips')
df
corr =df.corr(numeric_only = True)
sn.heatmap(corr , annot = True)
```

### Cluster Map (clustermap())

```python
#  cluster -> headmap plotation and clustering of allgomative

sn.clustermap(corr , annot = True)
```

### Styling & Themes

```python
z = np.random.rand(10 , 10)
fig = go.Figure(data =[
      go.Surface(z=z)
        ]
)
fig.show()
```

## Pivot table using heatmap

```python
flights = sn.load_dataset('flights')
flights
```

```python
flights = flights.pivot_table(values = 'passengers' , index = 'month' , columns = 'year')
```

```python
flights
```

```python
sn.heatmap(flights)
```

```python
lmplot = sn.lmplot(data = flights , x = 'passengers' , y= 'year' , hue = 'month' , markers= ['o', 'v', '^', 's', 'd', 'p', 'h', 'x', '+', '*', '<', '>'], palette= 'rainbow')
lmplot
```
